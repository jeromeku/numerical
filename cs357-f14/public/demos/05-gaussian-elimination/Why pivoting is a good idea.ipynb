# Why pivoting is a good idea

```python
import numpy as np
import numpy.linalg as la
```

```python

A = np.array([
    [1e-20, 1],
    [1, 1],
])
```

Now find an elimination matrix and go to town:

```python
U = A.copy()
```

```python
M = np.eye(2)
M[1,0] = -A[1,0]/A[0,0]
```

```python
U = M.dot(U)
U
```

```python
M
```

Now define `L`:

```python
L = la.inv(M)
```

```python
L.dot(U) - A
```
