# "Complete" QR vs Reduced QR

```python
import numpy as np
import numpy.linalg as la
```

```python
A = np.random.randn(20, 5)
```

```python
Q, R = la.qr(A)
```

```python
print(Q.shape)
print(R.shape)
```

```python
la.norm(Q.dot(R) - A)
```

Now try adding `"complete"` as an argument to `qr()`.

---------------
Let's make our own 'complete' QR.

```python
m, n = A.shape

Qnew = np.zeros((m,m))
Qnew[:m,:n] = Q
```

```python
for i in range(n, m):
    q = np.random.randn(m)
    for j in range(i):
        prev_q = Qnew[:, j]
        q = q - prev_q.dot(q)*prev_q
    q = q/la.norm(q)
    Qnew[:, i] = q
```

```python
Rnew = np.zeros((m,n))
Rnew[:n,:n] = R
```

```python
la.norm(
    Qnew.T.dot(Qnew) - np.eye(m))
```

```python
la.norm(
    Qnew.dot(Rnew) - A)
```
