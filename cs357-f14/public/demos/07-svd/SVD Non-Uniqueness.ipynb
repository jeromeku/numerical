```python
import numpy as np
import numpy.linalg as la
```

```python
X = np.random.randn(3,3)
```

```python
U, singvals, VT = la.svd(X)
V = VT.T
Sigma = np.diag(singvals)
```

```python
print(U)
print(Sigma)
print(V)
```

```python
X - U.dot(Sigma).dot(V.T)
```

```python
U2 = U.copy()
U2[:, 0] *= -1
V2 = V.copy()
V2[:, 0] *= -1
```

```python
X - U2.dot(Sigma).dot(V2.T)
```
