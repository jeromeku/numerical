```python
from __future__ import division

import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pt
```

```python
mesh = np.linspace(0, 1, 200)
```

```python
for i in range(10):
    pt.plot(mesh, mesh**i)
```

### Vandermonde matrix

```python
n = 5    # also try: 10, 20

V = np.array([mesh**i for i in range(n)]).T        
```

```python
la.cond(V)
```
